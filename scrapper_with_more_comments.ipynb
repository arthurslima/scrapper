```python
#uploading the CSV with all the URLs we want to look for
urls = pd.read_csv('urls.csv')
urls['winery'] = ''
urls['name'] = ''
urls['types'] = ''
urls['region'] = ''
urls['country'] = ''
urls['rating'] = ''
urls['n_rating'] ='' 
urls['price'] = ''
urls['review'] = ''
urls['grapes'] = ''
urls['regional'] = ''
urls['food'] = ''
print(urls.head())
```

```python
for i in range(0,len(urls)):
# for i in range(0,10):
    try:
        browser = webdriver.Chrome('/Users/arthurlima/Downloads/chromedriver 2')
        browser.get(urls['url'][i])

        #expanding the comments
        browser.find_element_by_xpath('//*[@id="all_reviews"]/div/div[2]/div/a').click()
        browser.find_element_by_xpath('//*[@id="all_reviews"]/div/div[2]/div/a').click()
        browser.find_element_by_xpath('//*[@id="all_reviews"]/div/div[2]/div/a').click()
        browser.find_element_by_xpath('//*[@id="all_reviews"]/div/div[2]/div/a').click()
        browser.find_element_by_xpath('//*[@id="all_reviews"]/div/div[2]/div/a').click()
        browser.find_element_by_xpath('//*[@id="all_reviews"]/div/div[2]/div/a').click()
        browser.find_element_by_xpath('//*[@id="all_reviews"]/div/div[2]/div/a').click()
        browser.find_element_by_xpath('//*[@id="all_reviews"]/div/div[2]/div/a').click()

        time.sleep(1) 

        #fetching some reviews:
        review = browser.find_element_by_xpath('//*[@id="all_reviews"]/div/div[2]/div').text

        browser.quit()

        # print('reviews: '+review)
        urls['review'][i] = review
        
        if i in (10,100,200,300,400,500,600,700,800,900,1000,1100,1200,1300,1400,1500,1600,1700,1800,1900,2000\
                ,2100,2200,2300,2400,2500,2600,2700,2800,2900,3000):
            urls.to_csv('result1.csv')
    except:
        pass
```
